## Questions & Answers

### Advertising Follow Up

1. Why did we decide to scale the inputs?

We wanted to infer variable importance by comparing the coefficients. Therefore, the variables need to share the same scale.

2. We observed a positive coefficient for TV but a negative for TV^2.
   What does this mean?

We introduced additional polynomial features to increase the flexibility of the model. TV^2 is one such feature (variable).

The model demonstrates that the effect of TV can be better approximated when using a combination of different terms.

When studying the sales-TV-scatterplot, we observe some kind of saturation for larger values of TV; the negative coefficient reflects for that.

The strength of the contribution is given by the absolute value of the coefficient, the direction by the sign.

---

It is not necessarily the case that coefficients can be interpreted. Naturally, the coefficients depend on the features available.

3. Using one of the models, describe and interpret the result of spending
   - no money
   - and an ever increasing amount of money
   
   on sales. Is this behaviour reasonable?

The models take a form like

$$\textrm{sales} = \beta_0 + \beta_1\textrm{TV} + \beta_2 \textrm{radio} + \beta_3 \textrm{newspaper} + \beta_4 \textrm{TV}\textrm{radio} + ...$$

- **no spending**:
  - $\textrm{TV} = \textrm{radio} = \textrm{newspaper} = 0$ and hence $\textrm{sales} = \beta_0$
  - We sell some goods even without advertising, that sounds reasonable.
- **every increasing spending**:
  - $\textrm{TV} \to \infty$ or $\textrm{radio} \to \infty$ or $\textrm{newspaper} \to \infty$ and hence $\textrm{sales} \to \pm\infty$
  - This behavior is not reasonable, cf. concept of diminishing returns.

### Parameter Optimization II
1. What does the decision region plot tell you?

```python
## given data
X, y = make_moons(n_samples=200, noise=0.3, random_state=123)

fig, ax = plt.subplots()
ax.scatter(X[:, 0], X[:, 1], c=y)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
```

```python
# train test split for demo purposes
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# model fit for demo purposes
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
# check score
model.score(X_train, y_train), model.score(X_test, y_test), 
```

```python
# example
fig, ax = plt.subplots()
plot_decision_regions(X_test, y_test, model, ax=ax)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
```

```python
# a different picture
from sklearn.tree import plot_tree

fig, ax = plt.subplots(figsize=(10, 20))
_ = plot_tree(model, ax=ax, filled=True, feature_names=['x1', 'x2'], impurity=False)
```

1. What does the decision region plot tell you?

The decision region plot explains how the classifier segments the input space, 
i.e. which regions correspond to which model decision.

```python
## similar to above
# use GridSearchCV to optimize the model parameters max_depth and min_samples_split
grid_search_cv = GridSearchCV(DecisionTreeClassifier(), 
                              {'max_depth': [1, 2, 3, 4, 5], 
                               'min_samples_split': [2, 4, 5, 10, 20]}, 
                              cv=10)
grid_search_cv.fit(X, y)
```

1. What is the optimal parameter setting you obtain for the `DecisionTreeClassifier` after performing a cross-validated grid search?

```python
# retrieve best parameter combination as
grid_search_cv.best_params_
```

```python
# fit a new model on the training data using the best parameters from above
best_model = grid_search_cv.best_estimator_
best_model.fit(X_train, y_train)
```

```python
# check scores
best_model.score(X_train, y_train), best_model.score(X_test, y_test) 
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(10, 6))
ax = axs[0]
plot_decision_regions(X_test, y_test, model, ax=ax)
ax.set_title('Default model')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax = axs[1]
plot_decision_regions(X_test, y_test, best_model, ax=ax)
ax.set_title('CV model')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
```

```python
fig, ax = plt.subplots(figsize=(10, 20))
_ = plot_tree(best_model, ax=ax, filled=True, feature_names=['x1', 'x2'], impurity=False)
```

1. How do the decision regions differ between the default and the cross-validated model?

Less fragmented, fewer details, coarser boundary.
Maybe too coarse.

```python
from sklearn.ensemble import RandomForestClassifier
# model fit for demo purposes
model = RandomForestClassifier()
model.fit(X_train, y_train)
print(model.score(X_train, y_train), model.score(X_test, y_test))
# example
fig, axs = plt.subplots(nrows=2, figsize=(6, 10))
ax = axs[0]
plot_decision_regions(X_train, y_train, model, ax=ax)
ax.set_title('Train')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax = axs[1]
plot_decision_regions(X_test, y_test, model, ax=ax)
ax.set_title('Test')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
```

1. Bonus: How does the decision region differ when using a `RandomForestClassifier`? Any ideas why it looks different?

Boundary appears smoother, yet allows for even smaller details. Overall structure seems to be captured well.

A `RandomForestClassifier` aggregates many different decision trees with each tree being trained on a randomly selected subset of the data thereby correcting for a single tree's behavior of overfitting.
